```python
import matplotlib.pyplot as plt
withmusic = [304,257,174,214,69,317,387,47,157,0,332,308,317,286,236,299,206,278,188,43,0,0,0,0,0]
withoutmusic = [292,270,47,288,524,292,364,316,287,75,282,149,274,319,213,3,324,2,128,219,94,164,0,0,0]
plt.plot(withmusic, [0] * len(withmusic), 'o', color='green', label='With Music')
plt.plot(withoutmusic, [1] * len(withoutmusic), 'o', color='orange', label='Without Music')
plt.title('Growth with and without music')
plt.yticks([0,1], ['With music', 'Without music'])
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
withmusic = [304,257,174,214,69,317,387,47,157,0,332,308,317,286,236,299,206,278,188,43,0,0,0,0,0]
withoutmusic = [292,270,47,288,524,292,364,316,287,75,282,149,274,319,213,3,324,2,128,219,94,164,0,0,0]
plt.hist(withmusic, bins=10, edgecolor='red', color='orange', alpha=0.6, label='with music')
plt.hist(withoutmusic, bins=10, edgecolor='red', color='blue', alpha=0.6, label='without music')
plt.title('Histogram of the growth with both music and without music')
plt.xlabel('Growth')
plt.ylabel('Frequency')
plt.legend(loc='best')
plt.show()
```
